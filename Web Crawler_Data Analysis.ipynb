```python
from selenium import webdriver
import time
from selenium.webdriver.common.action_chains import ActionChains #move_to_element
import re


#設定driver
path = 'C:\python/chromedriver.exe'
driver = webdriver.Chrome(path)

driver.get('https://shopee.tw/') #到蝦皮網址
time.sleep(1)

#輸入要搜尋的商品
print('輸入商品名稱：')
item = str(input())

#關閉廣告
ads_btn = driver.find_element_by_css_selector('div.shopee-popup__close-btn')
ads_btn.click()

#搜尋商品
t_field = driver.find_element_by_css_selector('input.shopee-searchbar-input__input') #搜尋欄
t_field.send_keys(item)
time.sleep(1)
t_btn = driver.find_element_by_css_selector('button.btn.btn-solid-primary.btn--s.btn--inline') #搜尋按鈕
t_btn.click()

#抓資料
time.sleep(3)
#查詢的資料
dataArr = []
    

def getData():
    index = 0 #第幾個商品
    while(1):
        productArr = []
        try:
            product = driver.find_elements_by_css_selector('div.fBhek2._2xt0JJ')[index] #視窗內的第index個商品
            ActionChains(driver).move_to_element(product).perform() #移動到元素位置


            product_name = product.find_element_by_css_selector('div.yQmmFK._1POlWt._36CEnF').text
            product_price = product.find_element_by_css_selector('div._32hnQt').text
            product_price_r = product_price.split('\n')
            product_sold = product.find_element_by_css_selector('div.go5yPW').text.replace('已售出','')
            
            #print(product_name)

            #if(len(product_price_r) == 1):
                #print(product_price_r[0], end='\n\n')
            #else:
                #print(product_price_r[0] + '-' + product_price_r[1])

            #print(product_sold,end='\n\n' )
            productArr.append(index)
            productArr.append(product_name)
            productArr.append(product_price_r)
            product_sold_new = wordgenerate(product_sold)
            productArr.append(product_sold_new)
            dataArr.append(productArr)
            
            time.sleep(0.1)
            index += 1

        except IndexError:
            break
    print(dataArr)

def wordgenerate(sold):
    if re.findall('萬', sold):
        ger_sold = eval(sold.replace('萬',''))*10000
        return '%d'% ger_sold
        
    else:
        ger_sold2 = sold.replace(',','')
        return ger_sold2
        
#開始
getData()
```

### 將資料依銷售量分為三個labels

```python
def ConvertLabel(score):
  if score >= 1000:
    return 'Good'
  
  elif 999 >= score >= 500:
    return 'Not Bad'
  
  else:
    return 'Bad'

def Convert(data):
  data['label'] = data['ProductSold'].map(lambda e : ConvertLabel(e))
  data = data[data['label'].isin(['Good','Not Bad','Bad'])]
  #data['label'] = data['label'].replace({'Good':2, 'Not Bad':1, 'Bad':0})
  return data

Convert(df)
```

### 運用jieba套件做分詞

```python
import jieba.posseg as pseg

def jieba_tokenizer(text):
  words = pseg.cut(text)
  return ' '.join([word for word,flag in words if flag != 'x'])

df['Product_tokenized'] = df.loc[:,'ProductName'].apply(jieba_tokenizer)

print(df.iloc[:,[0,5]])
```

### 將文本轉成數字序列

```python
from keras.preprocessing.text import Tokenizer

MAX_NUM_WORDS = 1000
tokenizer = Tokenizer(num_words = MAX_NUM_WORDS)

sentences = df['Product_tokenized'].apply(str).tolist()
tokenizer.fit_on_texts(sentences)
x_train = tokenizer.texts_to_sequences(sentences)
print(len(x_train))
print(x_train[:1])
```

```python
for seq in x_train[:9]:
  print([tokenizer.index_word[idx] for idx in seq])
```

### 序列的 Zero Padding

```python
#將所有的序列長度都設為30

import keras
from keras.preprocessing.sequence import pad_sequences
MAX_SEQUENCE_LENGTH = 30
x_train = keras.preprocessing.sequence.pad_sequences(x_train, maxlen = MAX_SEQUENCE_LENGTH)

print(x_train[0])
```

```python
for seq in x_train:
  assert len(seq) == 30
print('所有標題的序列長度皆為30!')
```

## label的處理

```python

import numpy as np

#定義每一個分類對應到的數字
label_to_index = {
    'Good': 2,
    'Not Bad': 1,
    'Bad': 0
}

#將分類標籤對應到剛定義的數字

y_train = df.label.apply(lambda x: label_to_index[x])

y_train = np.asarray(y_train).astype('float32')

print(y_train[:5])
```

### 做 one hot encoding

```python
from keras.utils import np_utils

y_train = np_utils.to_categorical(y_train)

print(y_train[:5])
```

```python
#用 tokenizer 裡頭的字典 index_word 還原文本看看：
for i, seq in enumerate(x_train[:5]):
  print(f'商品名稱{i + 1}:')
  print([tokenizer.index_word.get(idx,'')for idx in seq])
  print()
```

### 切割訓練資料集 & 驗證資料集

```python
from sklearn.model_selection import train_test_split

Validation_ratio = 0.2
Random_State = 50

x_train, x_val ,y_train, y_val = train_test_split(
    x_train,y_train,
    test_size = Validation_ratio,
    random_state = Random_State
)

print("Training Set")
print("-" * 10)
print(f"x_train: {x_train.shape}")
print(f"y_train : {y_train.shape}")

print("-" * 10)
print(f"x_val:   {x_val.shape}")
print(f"y_val :   {y_val.shape}")
print("-" * 10)
print("Test Set")
```

### 將一個詞彙或句子轉換成一個實數詞向量（Vectors of real numbers）的技術被稱之為詞嵌入（Word Embedding）

```python
from keras import layers
from keras.layers import Embedding, LSTM, concatenate, Dense, Input
from keras.models import Model

# 基本參數設置，有幾個分類
NUM_CLASSES = 3

# 在語料庫裡有多少詞彙
MAX_NUM_WORDS = 1000

# 一個標題最長有幾個詞彙
MAX_SEQUENCE_LENGTH = 30

# 一個詞向量的維度
NUM_EMBEDDING_DIM = 256

# LSTM 輸出的向量維度
NUM_LSTM_UNITS = 128

x_input = Input(shape = (30, ), dtype = 'int32')

#詞嵌入層
embedding_layer = layers.Embedding(MAX_NUM_WORDS, NUM_EMBEDDING_DIM)
x_embedded = embedding_layer(x_input)

#LSTM層
lstm = LSTM(NUM_LSTM_UNITS)
x_output = lstm(x_embedded)

#全連接層
dense = Dense(
    units = NUM_CLASSES,
    activation = 'softmax')
predictions = dense(x_output)

#model
model = Model(
    inputs = x_input,
    outputs = predictions
)

model.summary()
```

```python
x_train[:10].shape
```

```python
model.compile(
    optimizer = 'rmsprop',
    loss = 'categorical_crossentropy',
    metrics = ['accuracy']
)

history = model.fit(
    x = x_train,
    y = y_train,
    batch_size = 10,
    epochs = 10,
    validation_data = (x_val, y_val),
    shuffle = True
)
```
